```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

### Load data

```python
df = pd.read_csv("../data/raw/student_data.csv")
print(df.head())
```

### Basic info

```python
print(df.info())
print(df.describe())
```

### Missing values check

```python
print(df.isnull().sum())
```

### Correlation heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

### Distribution of final scores

```python
sns.histplot(df["final_score"], kde=True)
plt.title("Final Score Distribution")
plt.show()
```

```python
# Create directories if they dont exist
os.makedirs("data/processed", exist_ok=True)
```

### Save processed data

```python
df.to_csv("data/processed/student_scores_processed.csv", index=False)
print(" Processed data saved at data/processed/student_scores_processed.csv")
```
