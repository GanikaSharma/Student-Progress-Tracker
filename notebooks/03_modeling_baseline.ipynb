```python
# 03_modeling_baseline.ipynb

import pandas as pd
import os
import joblib
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# ======================
# 1. Load Dataset
# ======================
df = pd.read_csv("data/processed/student_scores_processed.csv")

# Features & Target
X = df.drop(["student_id", "final_score", "feedback"], axis=1)
y = df["final_score"]

# ======================
# 2. Convert Target to Categories
# ======================
bins = [0, 50, 75, 100]
labels = ["Low", "Medium", "High"]
y_class = pd.cut(y, bins=bins, labels=labels, include_lowest=True)

# Drop rows with NaN (if any)
mask = ~y_class.isna() & ~X.isna().any(axis=1)
X = X[mask]
y_class = y_class[mask]

# ======================
# 3. Train-Test Split
# ======================
X_train, X_test, y_train, y_test = train_test_split(
    X, y_class, test_size=0.2, random_state=42, stratify=y_class
)

# ======================
# 4. Logistic Regression
# ======================
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train, y_train)
y_pred_lr = log_reg.predict(X_test)

print("ðŸ“Š Logistic Regression Results")
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))

# ======================
# 5. Random Forest
# ======================
rf_clf = RandomForestClassifier(random_state=42)
rf_clf.fit(X_train, y_train)
y_pred_rf = rf_clf.predict(X_test)

print("\nðŸŒ² Random Forest Results")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))

# ======================
# 6. Save Models
# ======================
os.makedirs("models", exist_ok=True)

joblib.dump(log_reg, "models/log_reg_model.pkl")
joblib.dump(rf_clf, "models/rf_model.pkl")

print("âœ… Models saved successfully in 'models/'")
```
