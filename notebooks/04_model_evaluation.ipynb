```python
# 04_model_evaluation.ipynb

import pandas as pd
import joblib
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# ======================
# 1. Load Dataset
# ======================
df = pd.read_csv("data/processed/student_scores_processed.csv")

# Drop ID and feedback (must match training preprocessing!)
X = df.drop(["student_id", "final_score", "feedback"], axis=1)
y = df["final_score"]

# ======================
# 2. Convert Target to Categories
# ======================
bins = [0, 50, 75, 100]
labels = ["Low", "Medium", "High"]
y_class = pd.cut(y, bins=bins, labels=labels, include_lowest=True)

# Drop rows with NaN (same as before)
mask = ~y_class.isna() & ~X.isna().any(axis=1)
X = X[mask]
y_class = y_class[mask]

# ======================
# 3. Train-Test Split
# ======================
X_train, X_test, y_train, y_test = train_test_split(
    X, y_class, test_size=0.2, random_state=42, stratify=y_class
)

# ======================
# 4. Load Trained Models
# ======================
log_reg = joblib.load("models/logistic_regression.pkl")
rf_clf = joblib.load("models/random_forest.pkl")

# ======================
# 5. Evaluate Models
# ======================
results = {}

# Logistic Regression
y_pred_lr = log_reg.predict(X_test)
results["Logistic Regression"] = accuracy_score(y_test, y_pred_lr)
print("\nðŸ“Š Logistic Regression Report:")
print(classification_report(y_test, y_pred_lr))

# Random Forest
y_pred_rf = rf_clf.predict(X_test)
results["Random Forest"] = accuracy_score(y_test, y_pred_rf)
print("\nðŸŒ² Random Forest Report:")
print(classification_report(y_test, y_pred_rf))

# ======================
# 6. Save Summary Results
# ======================
results_df = pd.DataFrame(list(results.items()), columns=["Model", "Accuracy"])
results_df.to_csv("models/model_evaluation_results.csv", index=False)

print("\nâœ… Model Evaluation Summary:")
print(results_df)
print("\nðŸ“‚ Results saved to models/model_evaluation_results.csv")
```
